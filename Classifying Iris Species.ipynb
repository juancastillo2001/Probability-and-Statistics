The random state is used to randomly use the data starting with 0 so that the program does not get used to the same data, 75% of the data is provided for training


```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
iris_dataset['data'], iris_dataset['target'], random_state=0)
```

# ***Machine learning model.***

In this part of the code it is used to create the prediction algorithm from a new position.

```
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
```

```
knn.fit(X_train, y_train)
```

Only two-dimensional arrays are entered in this procedure

```
X_new = np.array([[5, 2.9, 1, 0.2]])
print("X_new.shape:", X_new.shape)
```

Prediction given by the model of the flower family



```
prediction = knn.predict(X_new)
print("Prediction:", prediction)
print("Predicted target name:",
       iris_dataset['target_names'][prediction])
```

The numbers generated are the results obtained based on the Dataset used



```
y_pred = knn.predict(X_test)
print("Test set predictions:\n", y_pred)
```

These results are only to measure the effectiveness of the program in terms of its operation and effectiveness.


```
print("Test set score: {:.2f}".format(np.mean(y_pred == y_test)))
```

```
print("Test set score: {:.2f}".format(knn.score(X_test, y_test)))
```
